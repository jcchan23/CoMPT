```python
import numpy as np
import pandas as pd
import pickle as pkl
```

```python
argv1 = '13C'
argv2 = 'test'
```

```python
all_data = pkl.load(open('./source/data_'+ argv1 + '.pickle','rb'))
data = all_data[argv2 + '_df']
data['value'] = data['value'].apply(lambda x: x[0])   # delete the [] and keep the dict
data
```

```python
data_mol = data['rdmol'].values.tolist()
data_cs = data['value'].values.tolist()
print(len(data_mol),len(data_cs))
```

```python
from rdkit import Chem
for mol,cs in list(zip(data_mol,data_cs))[::-1]:
    try:
        Chem.SanitizeMol(mol)
    except:
        data_mol.remove(mol)
        data_cs.remove(cs)
        continue
    # remove molecular with dot in the smile
    if '.' in Chem.MolToSmiles(mol):
        data_mol.remove(mol)
        data_cs.remove(cs)
        continue
```

```python
# 1H train 10248, 1H test 2552
# 13C train 21481, 13C test 5378
print(len(data_mol),len(data_cs))
```

```python
# store as a pickle file
with open('./preprocess/graph_'+ argv1 + '_' + argv2 + '.pickle','wb') as fw:
    pkl.dump([data_mol,data_cs],fw)
    
with open(f'./preprocess/{argv1}_{argv2}_info.txt','w') as writer:
    writer.write(f'Size of {argv2} set: {len(data_mol)}')
```
